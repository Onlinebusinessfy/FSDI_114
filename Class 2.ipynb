## Stacks and Queues

```python
# Simplified implementation of Stack and Queue
# Simplified refers to the fact that these are using built-ins
# Note: These are intended to give us a guide on how we could build these from scratch

class Stack:
    def __init__(self):
        self.items=[]

    def push(self, element):
        self.items.append(element)

    def pop(self):
        return self.items.pop()

    # Nice to have methods
    def is_empty(self):
        return self.items==[]

    def size(self):
        return len(self.items)

    def peek(self):
        return self.items[len(self.items)-1]
        
class Queue:
    def __init__(self):
        self.items=[]

    def enqueue(self, element):
        self.item.insert(0, element)

    def dequeue(self):
        return self.item.pop()

    # Nice to have methods
    def is_empty(self):
        return self.items==[]

    def size(self):
        return len(self.items)

    def peek(self):
        return self.items[len(self.items)-1]
```

## Variations
Research and describe at least one variation of a stack (e.g., bounded stack) and one variation of a queue (e.g., circular queue). Explain how they differ from the basic versions.
A bounded stack is a stack with fixed maximum size or capacity. Once the stack reaches its capacity, no further elements can be added until an element is removed. It differs from a Basic Stack because it does not impose a size limit unless explicitly implemented. In contrast, a bounded stack enforces a constraint, potentially leading to a stack overflow error if an attempt is made to push beyond the capacity. 
The Circular Queue is a type of queue where the last position connects back to the first position, forming a circle. This structure allows the queue to reuse memory efficiently when elements are dequeued. It defers from Basic Queue because elements are added to the rear and removed from the front, but it can leave unused memory at the front once elements are dequeued, and a circular queue overcomes this by reusing the vacated space, ensuring the queue behaves as if it wraps around.

## Advanced Concepts
Research and explain what a deque (double-ended queue) is. How does it differ from a standard queue and a stack? Provide at least one example of its practical use.
A deque is a data structure that allows elements to be added or removed from both ends, it supports insertion and deletion of both the front and the rear. It combines the functionality of the queue and the stack. It differs from the queue and stack because it can use at the same time the principle LIFO and FIFO, which makes it insert and delete from the front or the rear of it. A really common example would be when having a Job Scheduling with Priority, because the tasks have different levels of priority, so they could be added at the front or rear based on the urgency.

```python
## Implement QueueII like we did StackII

class QueueII:
    class __Node:
        def __init__(self, datum):
            self.data=datum
            self.next=None

    def __init__(self):
        self.front=None
        self.rear=None

    def enqueue(self, element):
        new_node=self.__Node(element)
        if not self.rear:
            self.front=self.rear=new_node
        else:
            self.rear.next=new_node
            self.rear=new_node

    def dequeue(self):
        if not self.front:
            raise IndexError("The queue is empty!")
        datum = self.front.data
        self.front = self.front.next
        if not self.front:
            self.rear = None
        return datum

    def is_empty(self):
        return self.front==None

    def size(self):
        count=0
        current=self.front
        while current:
            count+=1
            current=current.next
        return count

    def peek(self):
        if not self.front:
            raise IndexError("The queue is empty")
        return self.front.data
```
